```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn.gaussian_process.kernels as ker

#get m pairs of training and test data (m-fold cross-validation)
def get_cv(A, m):
    n = len(A)
    l = [i*n//m for i in range(m)] + [n-1]
    return [(np.vstack([A[:l[i]], A[l[i+1]:]]), A[l[i]:l[i+1]]) for i in range(m)]

#Get data from file
X0 = np.genfromtxt('../data/winequality-white.csv',delimiter=';')[1:]
X0, y = X0[:,:-1], X0[:,-1]

m = 15
X0_cv, y_cv = get_cv(X0, m), get_cv(y[:,None], m)

#n = number of input data points
#d = number of input dimensions
n, d = X0.shape
```

```python
def test(f, **kwargs):
    tmps = [f(**kwargs).fit(X0_cv[i][0], y_cv[i][0].ravel()) for i in range(m)]
    mse = [sum(np.round(tmp.predict(X0_cv[i][1]) - y_cv[i][1].ravel())**2)/len(y_cv[i][1]) for i, tmp in zip(range(m), tmps)]
    return sum(mse)/m, tmps
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
k = ker.DotProduct()
#k = ker.RBF()
#k = ker.WhiteKernel()
#k = ker.Matern()
#k = ker.RationalQuadratic()
test(GaussianProcessRegressor, normalize_y=True, kernel=k, alpha=1e-7)
```
